```python
#sift
def list_primes(maxN):
  dictionary1 = {}
  rangen=[x for x in range(2,maxN+1,1)]
  for i in rangen:
    dictionary1[i] = None
  def find_minimum(d):
    list1=[]
    for i,j in d.items():
      if j == None: list1.append(i)
    return min(list1)

  while None in dictionary1.values():
    p = find_minimum(dictionary1)
    dictionary1[p] = True
    n = 2
    while p * n <= max(dictionary1.keys()):
      dictionary1[p*n] = False
      n+=1

  return_list = []
  for i,j in dictionary1.items():
      if j: return_list.append(i)
  return return_list

print(list_primes(10))
```

```python
def glist_primes(maxN):
  dictionary1 = {}
  rangen=[x for x in range(2,maxN+1,1)]
  for i in rangen:
    dictionary1[i] = None
  def find_minimum(d):
    list1=[]
    for i,j in d.items():
      if j == None: list1.append(i)
    return min(list1)

  while None in dictionary1.values():
    p = find_minimum(dictionary1)
    dictionary1[p] = True
    n = 2
    while p * n <= max(dictionary1.keys()):
      dictionary1[p*n] = False
      n+=1

  return_list = []
  for i,j in dictionary1.items():
      if j: yield i

for x in glist_primes(100):
  print(x)
```

```python
class Money:
  def __init__(self,pounds,shillings, pence):
    self.pounds = pounds
    self.shillings = shillings
    self.pence = pence
  def __str__(self):
    return f"You have {self.pounds} pounds, {self.shillings} shillings, and {self.pence} pence."
  def __repr__(self):
    return f"Money object in old english denominations, it contians {self.pounds} pounds, {self.shillings} shillings, and {self.pence} pence."
  def __add__(self,n):
    if type(n) == int:
      self.pence += n
      while self.pence > 20:
        self.shillings += 1
        self.pence += -20
      while self.shillings > 12:
        self.pounds += 1
        self.shillings += -12
    elif type(n) == Money:
      self.shillings = self.shillings + n.shillings
      while self.shillings > 20:
        self.pence += 1
        self.shillings += -20
      self.shillings = self.shillings + n.shillings
      while self.shillings > 12:
        self.pounds += 1
        self.shillings += -12
      self.pounds = self.pounds + n.pounds
    return self

money1 = Money(1,2,3)
print(money1)
money1 = money1 + 5
repr(money1)
money2 = Money(2,2,2)
repr(money1 + money2)
```
